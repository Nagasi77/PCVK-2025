```python
black1_img = np. zeros(shape=(512, 512,3), dtype=np.int16)
plt.imshow(black1_img)
```

```python
#koordinat
cv.rectangle(black1_img,pt1=(384,0),pt2=(510, 150),color=(0,255,0), thickness=10)
plt.imshow(black1_img)
```

```python
cv.rectangle(black1_img, pt1=(200,200), pt2=(300,300),color=(0,0,255), thickness=15)
plt.imshow(black1_img)
```

```python
cv.circle(black1_img,center=(100,100), radius=50, color=(255,0,0),thickness= 8)
plt.imshow(black1_img)
```

```python
cv.line(black1_img,pt1=(0,0),pt2=(512,512),color=(255,255,255),thickness=5)
plt.imshow(black1_img)
```

```python
font = cv.FONT_HERSHEY_SIMPLEX
cv.putText(black1_img, text="Hello", org=(10, 500), fontFace=font, fontScale=4, color=(255, 255, 0), thickness=2, lineType=cv.LINE_AA)
plt.imshow(black1_img)
```

```python
black_img2 = np.zeros(shape=(512, 512, 3), dtype=np.int32)
plt.imshow(black_img2)
```

```python
vertices = np.array([[100,300],[200,200],[400,300],[200,400]],dtype=np.int32)
vertices
```

```python
pts = vertices.reshape((-1,1,2)) # nilai 2 tiap titik 3 channel perwakilan R, G, dan B
pts
```

```python
cv.polylines(black_img2,[pts],isClosed=True,color=(255,0,0),thickness=5)
plt.imshow(black_img2)
```
